```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Send request
url = "https://quotes.toscrape.com"
response = requests.get(url)

# Step 2: Parse HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Step 3: Extract quotes, authors, and tags
quotes = soup.find_all('span', class_='text')
authors = soup.find_all('small', class_='author')
tags = soup.find_all('div', class_='tags')

quote_list = []
author_list = []
tag_list = []

for i in range(len(quotes)):
    quote_list.append(quotes[i].text)
    author_list.append(authors[i].text)
    tag_list.append(
        ", ".join([tag.text for tag in tags[i].find_all('a', class_='tag')])
    )

# Step 4: Convert extracted data into DataFrame
data = pd.DataFrame({
    'Quote': quote_list,
    'Author': author_list,
    'Tags': tag_list
})

# Step 5: Display collected data
print("Collected Data: \n")
print(data)
```
