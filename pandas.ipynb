```python
import pandas as pd

# a. Creating DataFrames
data1 = {
    'Name': ['Ravi', 'Sita', 'Kiran'],
    'Age': [25, 30, 22]
}

data2 = {
    'Name': ['Arun', 'Meena'],
    'Age': [28, 26]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print("DataFrame 1: \n", df1)
print("DataFrame 2: \n", df2)

# b. concat()
merged_df = pd.concat([df1, df2], ignore_index=True)
print("Merged DataFrame: \n", merged_df)

# c. Setting conditions
filtered_df = merged_df[merged_df['Age'] > 25]
print("Filtered DataFrame (Age > 25): \n", filtered_df)

# d. Adding a new column
merged_df['Status'] = ['Active', 'Active', 'Inactive', 'Active', 'Inactive']
print("DataFrame with New Column: \n", merged_df)
```

```python
import pandas as pd
import numpy as np

# Creating a DataFrame with NaN values
data = {
    'Name': ['Ravi', 'Sita', 'Kiran', 'Meena', np.nan],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [45000, 52000, 48000, 50000, 47000]
}

df = pd.DataFrame(data)
print("Original DataFrame: \n", df)

# a. Filling NaN with string
df_filled = df.fillna('Unknown')
print("DataFrame after filling NaN: \n", df_filled)

# b. Sorting based on column values
df_sorted = df_filled.sort_values('Salary', ascending=True)
print("Sorted DataFrame by Salary: \n", df_sorted)

# c. groupby()
df_grouped = df_filled.groupby('Department').mean(numeric_only=True)
print("Grouped Data (Department-wise Average Salary): \n", df_grouped)
```
